```python
import pandas as pd
from pathlib import Path
from sklearn.metrics import roc_auc_score, average_precision_score
import numpy as np

split = "validation"
output_dir = "full_pairs_within_split"
split_path = Path(output_dir) / f"split_{split}"

all_files = sorted(split_path.glob("batch_*.parquet"))

y_true_list = []
y_score_list = []

for f in all_files:
    df = pd.read_parquet(f, columns=["label", "score"])  # читаем только нужные столбцы
    y_true_list.append(df["label"].astype(int).to_numpy())
    y_score_list.append(df["score"].astype(float).to_numpy())
    del df

y_true = np.concatenate(y_true_list)
y_score = np.concatenate(y_score_list)

roc_auc = roc_auc_score(y_true, y_score)
pr_auc = average_precision_score(y_true, y_score)

print("ROC-AUC:", roc_auc)
print("PR-AUC (Average Precision):", pr_auc)
```
